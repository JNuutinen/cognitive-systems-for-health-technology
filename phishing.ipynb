```python
import pandas as pd
import numpy as np

from sklearn import tree
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report

import graphviz
from IPython.display import IFrame
```

```python
df = pd.read_csv("phishing.csv", sep=';')
df.describe()
```

```python
colnames = df.columns.get_values()
x = df.loc[:, "having_IP_Address":"Statistical_report"]
y = df.loc[:, "Result"]
classifier = tree.DecisionTreeClassifier(max_depth=3)
classifier.fit(x, y)
```

```python
dot_data = tree.export_graphviz(classifier, out_file=None, feature_names=colnames[:-1], class_names=['no','yes'])
graph = graphviz.Source(dot_data) 
graph.render("Phishing")

IFrame("Phishing.pdf", width=900, height=700)
```

```python
# predict
y_pred = classifier.predict(x)

# output confusion matrix
cm = confusion_matrix(y, y_pred)
print("Confusion matrix:\n",cm)

accuracy = (cm[0][0]+cm[1][1])/(cm[0][0]+cm[1][1]+cm[0][1]+cm[1][0])
print("Accuracy calculated from the training set = %.3f" % (accuracy))

print(classification_report(y, y_pred, target_names=['no', 'yes']))
```
